```python
import codecs
aa = codecs.open("output/run/d3ba_8_0.2_20_0.95/test_beam/pathscorrect", "r", encoding="utf-8").read()
bb = codecs.open("output/run/d3ba_8_0.2_20_0.95/test_beam/pathswrong", "r", encoding="utf-8").read()

with open('path_pred.txt', 'w') as fout:
    fout.write(aa)
    fout.write(bb)
```

```python
import json
import pickle
import h5py
import os
import glob
import shutil
import csv
from networkx.readwrite import json_graph

class DataIO(object):
    def __init__(self, data):
        self.data = data

    @staticmethod
    def is_fileexits(filename):
        if not os.path.isfile(filename):
            raise FileNotFoundError(filename)

    @classmethod
    def load_list_from_text(cls, filename):
        cls.is_fileexits(filename)
        triples = []
        with open(filename, 'r') as f:
            data = csv.reader(f, delimiter='\t')
            for line in data:
                triples.append(line)
        return triples
```

```python
questions = DataIO.load_list_from_text('path_pred.txt')
val_txt =  DataIO.load_list_from_text('codes/preprocessing/scenegraphs/test.txt')
with open('prediction.txt', 'w') as fout:

## 1339_Are there napkins under the utensil to the left of the rice?	TRUE

    for l_q in questions:
        ll = []
        qs_id = []
        qs_pred = []
        jj = l_q[0].split('_')
        #print(jj)

        for l_v in val_txt:
            if int(jj[0]) == int(l_v[0]):
                if jj[1] == l_v[2]:

                    if l_q[1]=='TRUE':
                        fout.write('{} {}\n'.format(int(l_v[1]),'yes'))
                    elif l_q[1]=='FALSE':
                        fout.write('{} {}\n'.format(int(l_v[1]),'no'))
                else:
                    continue
## prediction.txt will have format question_id answer e.g. 18503733 no
```

```python
import csv
main_dict = []
with open('prediction.txt') as fn:
    data = csv.reader(fn, delimiter=' ')

    #main_dict = dict.fromkeys(['questionId','prediction'])
    for line in data:
        minor_dict = {}
        #minor_dict = dict.fromkeys(['questionId','prediction'])
        minor_dict['questionId'] = line[0]
        minor_dict['prediction'] = line[1]
        main_dict.append(minor_dict)
```

```python
import json

json.dump(main_dict, open("prediction_1.json", 'w'))
```
